#Breast Cancer Classification with a simple Neural Network

##Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

##Data Collection and Processing

```python
#Loading the Data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#Loading the Data to a DataFrame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
data_frame.head()
```

```python
#Adding the 'target' column to the DataFrame
data_frame['label'] = breast_cancer_dataset.target
data_frame.tail()
```

```python
#Number of Rows and Columns
data_frame.shape
```

```python
#Getting some information about the Data
data_frame.info()
```

```python
#Checking the missing values
data_frame.isnull().sum()
```

```python
#Statistical Measures about the Data
data_frame.describe()
```

```python
#Checking the Distribution of the Target Variable
data_frame['label'].value_counts()
```

1 -> Benign

0 -> *Malignant*

```python
data_frame.groupby('label').mean()
```

##Separating the Features and Target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

##Standardize the Data


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

##Building the Neural Network

```python
#Importing TensorFlow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#Setting up the Layers of NN
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30, )),  #Input Layer - 30 neurons (= number of features)
    keras.layers.Dense(20, activation='relu'),  #Hidden Layer - 20 neurons
    keras.layers.Dense(2, activation='sigmoid')  #Output Layer - 2 neurons (= number of classes(0 and 1))
])
```

```python
#Compiling the NN
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#Training the NN
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

##Visualizing Accuracy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training Data', 'Validation Data'], loc='lower right');
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training Data', 'Validation Data'], loc='upper right');
```

##Accuracy of the Model on Test Data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
print(loss)
```

```python
#Converting the Prediction Probability to Class Labels
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

##Building the Predictive System

```python
input_data = (17.99,10.38,122.8,1001,0.1184,0.2776,0.3001,0.1471,0.2419,0.07871,1.095,0.9053,8.589,153.4,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.6,2019,0.1622,0.6656,0.7119,0.2654,0.4601,0.1189) #Put Data Point from the DataSet

#Change the Input Data to NumPy Array
input_data_as_numpy_array = np.asarray(input_data)

#Reshape the Array as we are predicting for 1 Data Point
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

#Standardizing the Input Data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The Tumor is Malignant')
else:
  print('The Tumor is Benign')
```

Which is correct prediction according to the DataSet
